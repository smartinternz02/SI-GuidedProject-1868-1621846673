# NumPy Exercises

#### Import NumPy as np

```python
import numpy as np
```

### Create an array of 10 zeros 

```python
np.zeros(10,dtype=int)
```

#### Create an array of 10 ones

```python
np.ones(10,dtype=int)
```

#### Create an array of 10 fives

```python
np.full(10,5)
```

#### Create an array of the integers from 10 to 50

```python
np.arange(10,51)
```

#### Create an array of all the even integers from 10 to 50

```python
np.arange(10,51,2)
```

#### Create a 3x3 matrix with values ranging from 0 to 8

```python
np.arange(0,9).reshape(3,3)
```

#### Create a 3x3 identity matrix

```python
np.identity(3,dtype=int)
```

#### Use NumPy to generate a random number between 0 and 1

```python
np.random.rand()
```

#### Use NumPy to generate an array of 25 random numbers sampled from a standard normal distribution

# Seaborn visualization on Iris Dataset

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset("iris")
```

```python
plt.scatter(df['sepal_width'],df['sepal_length'])
```

```python
plt.scatter(df['petal_width'],df['petal_length'])
```

```python
df.plot(x="sepal_width",y="sepal_length",kind="bar")
```

```python
df.plot(x="petal_width",y="petal_length",kind="bar")
```

```python
sns.relplot(data=df,x="sepal_length",y="sepal_width",hue="species")
```

```python
sns.distplot(df.sepal_length,kde=True)
```

```python
sns.distplot(df.sepal_width,kde=True)
```

```python
sns.distplot(df.petal_width,kde=True)
```

```python
sns.distplot(df.petal_length,kde=True,bins=10)
```

```python
sns.countplot(data=df,x="species")
```

```python
sns.jointplot(data=df,x="sepal_width",y="sepal_length")
```

```python
sns.jointplot(data=df,x="petal_width",y="petal_length")
```

```python
sns.boxplot(x="species",y="sepal_width",data=df)
```

```python
sns.boxplot(x="species",y="sepal_length",data=df)
```

```python
sns.boxplot(x="species",y="petal_width",data=df)
```

```python
sns.boxplot(x="species",y="petal_length",data=df)
```

```python
sns.heatmap(df.corr(),annot=True,cmap="YlGnBu")
```

```python
sns.pairplot(data=df,hue="species",kind="scatter",palette="husl")
```
